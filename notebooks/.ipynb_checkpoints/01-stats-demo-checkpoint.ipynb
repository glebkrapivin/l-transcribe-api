```python
import json
```

```python
x = json.load(open('../data/zagit/20230204/20230204.json'))
```

```python
s= x["results"][-1]["alternatives"][0]["words"]
```

```python
s[0]
```

```python
speaker = 2
st = ""
for item in s:
    if item["speakerTag"]!=speaker:
        print(f'Speaker {speaker}: {st}')
        speaker = item["speakerTag"]
        st = ""
    st  +=  item["word"] + ' '
        
```

```python
x["results"][0]["alternatives"][0]
```

```python
x["results"] = x["results"][:-1]
```

```python
time_talking = 0 
for result in x["results"]:
    words = result["alternatives"][0]["words"]
    
    for word in words: 
        time_talking += float(word["endTime"].replace('s', '')) - float(word["startTime"].replace('s', ''))
    
```
